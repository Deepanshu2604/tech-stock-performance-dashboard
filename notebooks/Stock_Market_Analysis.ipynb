## Data Cleaning & Sorting

### Date column ko datetime me convert 

```python
df["Date"]=pd.to_datetime(df["Date"])
```

### Sorted Data

```python
df = df.sort_values(by=['Ticker', 'Date'])
```

```python
df
```

### Index Reset

```python
df=df.reset_index(drop=True)
```

```python
df.head()
```

```python
df.info()
```

### Daily Returns Calculate

```python
df["Daily Return"]=df.groupby("Ticker")["Close"].pct_change()

# pct_change() → percentage change calculate karta hai
```

```python
df[['Ticker', 'Date', 'Close', 'Daily_Return']].head(10)
```

```python
df["Daily_Return"].describe()
```

## Moving Average (20-Day) calculate 

```python
df["MA_20"]=(df.groupby("Ticker")["Close"].transform(lambda x:x.rolling(window=20).mean()))

# rolling(window=20) → last 20 trading days
```

```python
df[["Ticker","Date","Close","MA_20"]].head(25)
```

```python
sample=df[df["Ticker"]==df["Ticker"].unique()[0]]
plt.plot(sample["Date"],sample["Close"],label="Close Price")
plt.plot(sample["Date"],sample["MA_20"],label="20-Day MA")
plt.legend()
plt.show()
```

## Volatility Calculate

#### High volatility → zyada risk 
#### Low volatility → zyada stable stock

```python
volatility_df=(
    df.groupby("Ticker")["Daily_Return"]
    .std()
    .reset_index()
    .rename(columns={"Daily_Return":"Volatility"})
)
```

```python
volatility_df
```

```python
c=["k","g","c","r"]
plt.bar(volatility_df["Ticker"],volatility_df["Volatility"],color=c)
plt.title('Stock Volatility Comparison')
plt.show()

# Insights: Netflix ka price daily zyada upar-neeche hota hai
#           Apple comparatively stable hai
```

## Correlation Calculate

```python
returns_pivot = df.pivot(
    index='Date',
    columns='Ticker',
    values='Daily_Return'
)
```

```python
correlation_matrix=returns_pivot.corr()
```

```python
correlation_matrix
```

```python

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between Stock Returns')
plt.show()
```
